```python
name=[]
category = []
price = []
colors = []
color_name=[]
total_sizes_list=[]
sizes = []
features=[]
comfort = []
product_code=[]
reviews=[]
ratings=[]

for product in products:
    driver.get(product)
    
    #Extracting Product code
    sku=product.split('/')
    product_code.append(sku[-1].split('.')[0])
    
    #Extracting Name
    title = driver.find_element(By.XPATH,'//div[@class="product-description___1TLpA"]/h1/span')
    name.append(title.text)
    
    #Extracting Price
    try:
        cost = driver.find_element(By.XPATH,'//div[@class="product-price___2Mip5 gl-vspace"]/div/div/div')
        
        #Find reducted Price
        try:
            reduced = cost.find_element(By.XPATH, './/div[@class="gl-price-item gl-price-item--sale notranslate"]')
            price.append(reduced.text.strip('₹'))
            
        except (StaleElementReferenceException, NoSuchElementException):
            try:
                original = cost.find_element(By.XPATH, './/div[@class="gl-price-item notranslate"]')
                price.append(original.text.strip('₹'))
            except NoSuchElementException:
                             price.append('NA')
    except (StaleElementReferenceException, NoSuchElementException):
                             price.append('NA')
            
    #Extracting Number of colors
    try:
        colr = driver.find_element(By.ID, 'available-colors-label')
        colors.append(colr.text.split(' ')[0])
    except (StaleElementReferenceException, NoSuchElementException):
        colors.append('NA')
        
    #Extracting the name of the colors
    try:
        colortag = driver.find_element(By.XPATH,'//div[@data-auto-id="color-label"]')
        color_name.append(colortag.text)
    except (StaleElementReferenceException, NoSuchElementException):
        color_name.append('NA')
        
    try:
        button= driver.find_element(By.XPATH, '//button[@data-auto-id="reload-availability"]')
        button.click()
        time.sleep(2)
    except NoSuchElementException:
        pass
    
    
    try:
        # Exrtracting count of size and Sizes
        total_size = driver.find_element(By.XPATH,'//div[@data-auto-id="size-selector"]')

        # Exrtracting count of size
        try:
            count_size = total_size.find_elements(By.XPATH,'//button[@class="gl-label size___2lbev"]/span')
            total_sizes_list.append(len(count_size))

            #Extracting sizes
            sizes_text = [i.text for i in count_size]
            sizes.append(sizes_text)

        except (StaleElementReferenceException, NoSuchElementException):
            total_sizes_list.append('NA')
            sizes.append('NA')
            
    except NoSuchElementException:
        total_sizes_list.append('Not Loading')
        sizes.append('Not Loading')

        
    # Scrolling:====
    scroll = driver.find_element(By.ID, 'navigation-target-description')
    driver.execute_script("arguments[0].scrollIntoView(true);", scroll)
    
    #Extrcting description
    button1 = driver.find_element(By.XPATH, '//div[@id="navigation-target-description"]/div')
    button1.click()
    time.sleep(2)
    # Feature:
    desc = driver.find_element(By.XPATH, '//div[@class="text-content___13aRm"]')
    features.append(desc.text)
    
     #extracting comfort
    button2 = driver.find_element(By.XPATH, '//div[@id="navigation-target-specifications"]')
    button2.click()
    
    comf= driver.find_elements(By.XPATH, '//div[@data-auto-id="specifications-section"]/ul/li')
    comfort_paragraph = ' '.join([li.text for li in comf])
    comfort.append(comfort_paragraph)
    time.sleep(2)
    
    #Extracting Category
    try:
        table = driver.find_element(By.XPATH, '//div[@class="gl-table gl-table--striped-columns"]/div')
        rows = table.find_elements(By.XPATH, './/div[@class="gl-table__row gl-table__row--body"]')  # Find all rows in the body
        for row in rows:
            cell = row.find_element(By.XPATH, './/div[@role="rowheader"]/div')
            if cell.text.strip() == 'Gender':
                value = row.find_element(By.XPATH, './/div[@role="cell"]/div')
                category.append(value.text)
                
    except NoSuchElementException:
        category.append('NA')
        
    #Scrolling till review
    try:
        scroll1 = driver.find_element(By.ID, 'navigation-target-reviews')
        driver.execute_script("arguments[0].scrollIntoView(true);", scroll1)
    except NoSuchElementException:
        pass
    
    #Extracting Number of reviews:
    try:
        rev=driver.find_element(By.XPATH,'//div[@id="navigation-target-reviews"]/div/button/div/div/h2')
        reviews.append(rev.text.split('(')[1].strip(')'))
        
    except NoSuchElementException:
        reviews.append('NA')
        
    #Extracting rating
    try:
        rat = driver.find_element(By.XPATH, '//div[@id="navigation-target-reviews"]/div/button/div/div/div/span')
        ratings.append(rat.text)
        
    except NoSuchElementException:
        ratings.append('NA')
```

```python
products=prd_url[2000:]


name3=[]
category3 = []
price3 = []
colors3 = []
color_name3=[]
total_sizes_list3=[]
sizes3 = []
features3=[]
comfort3 = []
product_code3=[]
reviews3=[]
ratings3=[]

for product in products:
    driver.get(product)
    
    #Extracting Product code
    sku=product.split('/')
    product_code3.append(sku[-1].split('.')[0])
    
    #Extracting Name
    try:
        title = driver.find_element(By.XPATH,'//div[@class="product-description___1TLpA"]/h1/span')
        name3.append(title.text)
        
    except NoSuchElementException:
        name3.append('NA')
    
    #Extracting Price
    try:
        cost = driver.find_element(By.XPATH,'//div[@class="product-price___2Mip5 gl-vspace"]/div/div/div')
        
        #Find reducted Price
        try:
            reduced = cost.find_element(By.XPATH, './/div[@class="gl-price-item gl-price-item--sale notranslate"]')
            price3.append(reduced.text.strip('₹'))
            
        except (StaleElementReferenceException, NoSuchElementException):
            try:
                original = cost.find_element(By.XPATH, './/div[@class="gl-price-item notranslate"]')
                price3.append(original.text.strip('₹'))
            except NoSuchElementException:
                             price3.append('NA')
                    
    except (StaleElementReferenceException, NoSuchElementException):
                             price3.append('NA')
            
    #Extracting Number of colors
    try:
        colr = driver.find_element(By.ID, 'available-colors-label')
        colors3.append(colr.text.split(' ')[0])
    except (StaleElementReferenceException, NoSuchElementException):
        colors3.append('NA')
        
    #Extracting the name of the colors
    try:
        colortag = driver.find_element(By.XPATH,'//div[@data-auto-id="color-label"]')
        color_name3.append(colortag.text)
    except (StaleElementReferenceException, NoSuchElementException):
        color_name3.append('NA')
        
    try:
        button= driver.find_element(By.XPATH, '//button[@data-auto-id="reload-availability"]')
        button.click()
        time.sleep(2)
    except NoSuchElementException:
        pass
    
    
    try:
        # Exrtracting count of size and Sizes
        total_size = driver.find_element(By.XPATH,'//div[@data-auto-id="size-selector"]')

        # Exrtracting count of size
        try:
            count_size = total_size.find_elements(By.XPATH,'//button[@class="gl-label size___2lbev"]/span')
            total_sizes_list3.append(len(count_size))

            #Extracting sizes
            sizes_text = [i.text for i in count_size]
            sizes3.append(sizes_text)

        except (StaleElementReferenceException, NoSuchElementException):
            total_sizes_list3.append('NA')
            sizes3.append('NA')
            
    except NoSuchElementException:
        total_sizes_list3.append('Not Loading')
        sizes3.append('Not Loading')

        
    # Scrolling:====
    try:
        scroll = driver.find_element(By.ID, 'navigation-target-description')
        driver.execute_script("arguments[0].scrollIntoView(true);", scroll)
    except NoSuchElementException:
        pass
    
    
    #Extrcting description
    try:
        button1 = driver.find_element(By.XPATH, '//div[@id="navigation-target-description"]/div')
        button1.click()
        time.sleep(2)
        
        # Feature:
        desc = driver.find_element(By.XPATH, '//div[@class="text-content___13aRm"]')
        features3.append(desc.text)
        
    except NoSuchElementException:
        features3.append('NA')
    
    
     #extracting comfort
    try:
        button2 =  driver.find_element(By.XPATH, '//div[@id="navigation-target-specifications"]')
        button2.click()

        comf= driver.find_elements(By.XPATH, '//div[@data-auto-id="specifications-section"]/ul/li')
        comfort_paragraph = ' '.join([li.text for li in comf])
        comfort3.append(comfort_paragraph)
        time.sleep(2)
        
    except NoSuchElementException:
        comfort3.append('NA')
    
    
    #Extracting Category
    try:
        table = driver.find_element(By.XPATH, '//div[@class="gl-table gl-table--striped-columns"]/div')
        rows = table.find_elements(By.XPATH, './/div[@class="gl-table__row gl-table__row--body"]')  # Find all rows in the body
        for row in rows:
            cell = row.find_element(By.XPATH, './/div[@role="rowheader"]/div')
            if cell.text.strip() == 'Gender':
                value = row.find_element(By.XPATH, './/div[@role="cell"]/div')
                category3.append(value.text)
                
    except NoSuchElementException:
        category3.append('NA')
     
    
    #Scrolling till review
    try:
        scroll1 = driver.find_element(By.ID, 'navigation-target-reviews')
        driver.execute_script("arguments[0].scrollIntoView(true);", scroll1)
    except NoSuchElementException:
        pass
    
    
    #Extracting Number of reviews:
    try:
        rev=driver.find_element(By.XPATH,'//div[@id="navigation-target-reviews"]/div/button/div/div/h2')
        reviews3.append(rev.text.split('(')[1].strip(')'))
        
    except NoSuchElementException:
        reviews3.append('NA')
     
    
    #Extracting rating
    try:
        rat = driver.find_element(By.XPATH, '//div[@id="navigation-target-reviews"]/div/button/div/div/div/span')
        ratings3.append(rat.text)
        
    except NoSuchElementException:
        ratings3.append('NA')
```
