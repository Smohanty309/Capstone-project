### 1. Creating Database Named `Shoe_brand`

```python
# Create a SQLite database
conn = sqlite3.connect('Shoe_Brand.db')
cursor = conn.cursor()
```

### 2. Creating 3 tables inside DataBase Shoe_Brand Named `data1`, `data2` and `data3`

```python
# Create data1 table
cursor.execute('''
    CREATE TABLE IF NOT EXISTS data1 (
        sno INTEGER PRIMARY KEY,
        Brand TEXT,
        Shoename TEXT,
        Category TEXT,
        no_of_colors INTEGER,
        Price REAL
    )
''')

# Create data2 table
cursor.execute('''
    CREATE TABLE IF NOT EXISTS data2 (
        sno INTEGER PRIMARY KEY,
        Count_of_sizes INTEGER,
        Name_of_colors TEXT,
        Product_code TEXT
    )
''')

# Create data3 table
cursor.execute('''
    CREATE TABLE IF NOT EXISTS data3 (
        sno INTEGER PRIMARY KEY,
        Review INTEGER,
        Size TEXT,
        Comfort TEXT,
        Features TEXT,
        star_rating FLOAT
    )
''')

# Commit the changes
conn.commit()


print("SQLite database 'Shoe_Brand' and tables 'data1,' 'data2,' and 'data3' created successfully.")
```

### 3. Uploading all the the 3 csv's inside their respected tables

```python
# Specify the paths to CSV files
data1_csv = 'Data1.csv'
data2_csv = 'Data2.csv'
data3_csv = 'Data3.csv'

# Read CSV files into DataFrames
data1_df = pd.read_csv(data1_csv)
data2_df = pd.read_csv(data2_csv)
data3_df = pd.read_csv(data3_csv)

# Uploading data to respective tables
data1_df.to_sql('data1', conn, if_exists='replace', index=False)
data2_df.to_sql('data2', conn, if_exists='replace', index=False)
data3_df.to_sql('data3', conn, if_exists='replace', index=False)

# Commit changes
conn.commit()

print("Data from CSV files uploaded to their respective tables in the 'Shoe_Brand' database.")
```

```python
# Write a query to find the cheapest women's shoe in a specific color (e.g., 'Black ').

query = '''
SELECT ShoeName, price
FROM Data1
WHERE Category = 'Women' AND (
    LOWER(no_of_colors) LIKE '%black%' OR
    LOWER(no_of_colors) LIKE '%/black%' OR
    LOWER(no_of_colors) LIKE '%black/%'
)
ORDER BY price DESC
LIMIT 1;
'''

# Execute the SQL query and store the result in a DataFrame
result_df = pd.read_sql_query(query, conn)

# Print the DataFrame
result_df
```

# 7 Join SQL Queries using all 3 tables

```python
# Write a query to find the top-rated men's shoes along with their sizes from "Table1" and "Table3."

query = '''
SELECT
    d1.Brand,
    d1.ShoeName,
    d1.Category,
    d1.no_of_colors,
    d1.Price,
    d3.Size,
    d3.star_rating
FROM
    data1 d1
JOIN
    data3 d3 ON d1.sno = d3.Review
WHERE
    d1.Category = 'Men'
ORDER BY
    d3.star_rating DESC
    LIMIT 5;
'''

# Execute the SQL query and store the result in a DataFrame
result_df = pd.read_sql_query(query, conn)

# Print the DataFrame
result_df
```

```python
# Write a query that calculates the average comfort rating for each category from "Table1" and "Table3."

query = '''
SELECT
    shoename,
    AVG(CAST(d3.Comfort AS FLOAT)) AS avg_comfort_rating
FROM
    data1 d1
JOIN
    data3 d3 ON d1.sno = d3.Review
GROUP BY
    d1.Category;

'''

# Execute the SQL query and store the result in a DataFrame
result_df = pd.read_sql_query(query, conn)

# Print the DataFrame
result_df
```

```python
# Write a query that identifies products with a durability/quality/performance rating higher than the average from "Table1" and "Table3."

query = '''
WITH AvgRatings AS (
    SELECT AVG(star_rating) AS avg_star_rating
    FROM data3
    WHERE Review IS NOT NULL
)
SELECT
    d1.Brand,
    d1.ShoeName,
    d1.Category,
    d1.no_of_colors,
    d1.Price,
    d3.Size,
    d3.star_rating
FROM
    data1 d1
JOIN
    data3 d3 ON d1.sno = d3.Review
CROSS JOIN
    AvgRatings avg
WHERE
    d3.star_rating > avg.avg_star_rating
ORDER BY
    d3.star_rating DESC;
'''

# Execute the SQL query and store the result in a DataFrame
result_df = pd.read_sql_query(query, conn)

# Print the DataFrame
result_df
```

```python
# Write a subquery that finds products with comfort ratings above the average comfort rating using "Table1" and "Table3."

query = '''
SELECT *
FROM
    Data1 d1
JOIN
    Data3 d3 ON d1.sno = d3.sno
WHERE
    d3.Comfort > (
        SELECT AVG(CAST(Comfort AS FLOAT))
        FROM Data3
        WHERE Comfort IS NOT NULL
    )
    AND d3.STAR_RATING > (
        SELECT AVG(STAR_RATING)
        FROM Data3
        WHERE STAR_RATING IS NOT NULL
    );

'''
# Execute the SQL query and store the result in a DataFrame
result_df = pd.read_sql_query(query, conn)

# Print the DataFrame
result_df
```
