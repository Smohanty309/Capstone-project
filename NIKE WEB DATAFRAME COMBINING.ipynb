## LOADING ALL THE DATA FRAMES FOR COMBING NIKE SHOES

```python
Product = pd.read_csv('main_Nike_urls.csv') # Main Products
Product.shape
```

```python
# Colors Urls

clr1= pd.read_csv('Nike_clrurls1st_500.csv')
clr1.shape
```

```python
clr2 =pd.read_csv('Nike_clrurls2nd_1000.csv')
clr2.shape
```

```python
clr3 =pd.read_csv('Nike_clrurls3rd_1600.csv')
clr3.shape
```

```python
clr=clr1.append(clr2)
```

```python
new=clr.append(clr3)
new
```

```python
Product= Product.append(new)
Product
```

```python
Product=Product.drop('Product_url', axis=1)
Product.shape
```

```python
Product['Name'] = Product['Name'].str.replace('Nike', '', case=False).str.strip()
```

```python
Product['Brand']= 'Nike'
```

```python
def extract_gender(row):
    if 'Women' in row['Category']:
        return 'Women'
    elif 'Men' in row['Category']:
        return 'Men'
    else:
        return 'Unisex'
```

```python
import numpy as np
```

```python
Product['Category'] = np.where(Product['Category'].str.contains('Women'), 'Women',
                        np.where(Product['Category'].str.contains('Men'), 'Men', 'Unisex'))
```

```python
Product
```

```python
Product.to_csv('Nike_All_Shoes.csv', index=False)
```

### TABLE 1

```python
name = Product['Name'].tolist()
category = Product['Category'].tolist()
brand = Product['Brand'].tolist()
colors = Product['Colors'].tolist()
price = Product['Price'].tolist()
```

```python
nike_Table1 =pd.DataFrame({
    'Brand': brand,
    'Shoename': name,
    'Category' : category,
    'No.Of_colors': colors,
    'Price' : price,   
})
nike_Table1
```

```python
nike_Table1.to_csv('nike_Table1.csv',index=False)
```

### Table2

```python
count_sizes = Product['Count_Sizes'].tolist()
colors_name = Product['Color_Name'].tolist()
product_code = Product['product_Code'].tolist()
```

```python
nike_Table2 = pd.DataFrame({
    'Count_of_sizes' : count_sizes,
    'Name_ of_colors' : colors_name,
    'Product_code' : product_code,
})
nike_Table2
```

```python
nike_Table2.to_csv('nike_Table2.csv',index=False)
```

```python
review = Product['Review'].tolist()
sizes = Product['Sizes'].tolist()
comfort = Product['Comfort'].tolist()
features = Product['Features'].tolist()
Rating = Product['Rating'].tolist()
```
