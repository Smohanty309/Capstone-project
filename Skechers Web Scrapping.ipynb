```python
# Importing Libraries
import pandas as pd
import selenium          
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys

import warnings                                  
warnings.filterwarnings("ignore")

import time                                      
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service

#Importing required exceptions which needs to be handled
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException, TimeoutException

options = webdriver.ChromeOptions()
options.add_experimental_option("excludeSwitches", ['enable-automation'])
options.add_argument('--disable-blink-features=AutomationControlled')
options.add_argument(
    "user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36")
options.add_argument("--remote-debugging-port=9222")

driver = webdriver.Chrome(options=options)

# Let's Install automated chrome browser and connect it to web driver
driver = webdriver.Chrome(service=Service())

#maximise the window
driver.maximize_window()
```

```python
name=[]
category = []
price = []
colors = []
color_name=[]
count_sizes=[]
total_sizes = []
features=[]
comfort = []
product_code=[]
reviews=[]
ratings=[]

for url in prd_url:
    driver.get(url)
    
    # Extracting Name===================================
    title = driver.find_element(By.XPATH,'//h2[@class="product-name hidden-sm-down"]')
    name.append(title.text)
    
    # Extracting category===========================
    cate = driver.find_element(By.XPATH,'//div[@class="pl-0 hidden-sm-down product-gendername"]')
    category.append(cate.text)
    
    # Extracting price=====================================================
    cost = driver.find_element(By.XPATH,'//div[@class="d-none d-md-flex mt-0 align-items-md-center pdp-price-block"]/div/div/span/span[1]')
    price.append(cost.text.strip('â‚¹'))
    
    #Extracting number of colors=================================================================
    color_tags = driver.find_element(By.XPATH,'//div[@class="attributes pl-0"]/div[2]/div/div')
    try:
        count= len(color_tags.find_elements(By.XPATH,'.//button[@class="color-attribute"]'))
        colors.append(count)
    except NoSuchElementException:
        colors.append('NA')
        
    #Extracting name of the product_code====================================================================
    
    code=driver.find_element(By.XPATH,'//div[@class="d-flex justify-content-between align-items-center"]/div[2]/span[2]/span')
    product_code.append(code.text)
    
    #Extracting Color_name======================================================================================
    color_name.append(code.text.split('-')[1])
    
    # Extracting size and count of sizes=======================================================
    size_selector=driver.find_element(By.XPATH,'//div[@class="select-size"]')
    count= len(size_selector.find_elements(By.XPATH,'.//button[@class="size-select  "]'))
    count_sizes.append(count)
    
    # Sizes available
    avail_size= size_selector.find_elements(By.XPATH,'.//button[@class="size-select  "]')
    size_text=[i.text for i in avail_size]
    total_sizes.append(size_text)
    
    # extracting Review======================================================================================
    try:
        review_ratings =driver.find_element(By.XPATH,'//div[@class="pr-snippet-stars-reco-inline pr-snippet-standard"]/div[1]')
        
        #Extracting Reviews........................
        try:
            rev=review_ratings.find_element(By.XPATH,'.//div[@class="pr-snippet-read-and-write"]/span')
            reviews.append(rev.text.split(' ')[0])
        except NoSuchElementException:
            reviews.append('NA')
            
    except NoSuchElementException:
        reviews.append('NA')
        
    #Extracting Ratings
    try:
        rat= driver.find_element(By.XPATH,'//div[@class="pr-snippet-stars pr-snippet-stars-png "][1]/div[2][1]')
        ratings.append(rat.text)
    except NoSuchElementException:
        ratings.append('NA')
    
    #Extracting features==============================================
    
    try:
        featr = driver.find_element(By.XPATH,'//div[@id="collapsible-details-1"]/div[2]/ul')
        features.append(featr.text)
    except NoSuchElementException:
        features.append('NA')
        
    # Extracting Comfort=======================================================
    try:
        comf = driver.find_element(By.XPATH,'//div[@class="col-sm-12 col-md-8 col-lg-9 value content pl-0 details-page-long-description"]/div[1]')
        comfort.append(comf.text)
    except NoSuchElementException:
        comfort.append('NA')
```

```python
name=[]
category = []
price = []
colors = []
color_name=[]
count_sizes=[]
total_sizes = []
features=[]
comfort = []
product_code=[]
reviews=[]
ratings=[]

for url in urls:
    driver.get(url)
    
    # Extracting Name===================================
    try:
        title = driver.find_element(By.XPATH,'//h2[@class="product-name hidden-sm-down"]')
        name.append(title.text)
    except NoSuchElementException:
        ('NA')
    
    # Extracting category===========================
    try:
        cate = driver.find_element(By.XPATH,'//div[@class="pl-0 hidden-sm-down product-gendername"]')
        category.append(cate.text)
    except NoSuchElementException:
        ('NA')
    
    # Extracting price=====================================================
    try:
        cost = driver.find_element(By.XPATH,'//div[@class="d-none d-md-flex mt-0 align-items-md-center pdp-price-block"]/div/div/span/span[1]')
        price.append(cost.text.strip('â‚¹'))
    except NoSuchElementException:
        price.append('NA')
    
    #Extracting number of colors=================================================================
    
    color_tags = driver.find_element(By.XPATH,'//div[@class="attributes pl-0"]/div[2]/div/div')
    try:
        count= len(color_tags.find_elements(By.XPATH,'.//button[@class="color-attribute"]'))
        colors.append(count)
    except NoSuchElementException:
        colors.append('NA')

    #Extracting name of the product_code====================================================================

    code=driver.find_element(By.XPATH,'//div[@class="d-flex justify-content-between align-items-center"]/div[2]/span[2]/span')
    product_code.append(code.text)

    #Extracting Color_name======================================================================================
    elements = driver.find_elements(By.XPATH,'//button[@class="color-attribute"]/span[2]')
    for element in elements:
        if element.text.strip().lower() == 'selected':
            color_name.append(element.get_attribute('id'))
    

     # Extracting size and count of sizes=======================================================
    size_selector=driver.find_element(By.XPATH,'//div[@class="select-size"]')
    count= len(size_selector.find_elements(By.XPATH,'.//button[@class="size-select  "]'))
    count_sizes.append(count)

     # Sizes available
    avail_size= size_selector.find_elements(By.XPATH,'.//button[@class="size-select  "]')
    size_text=[i.text for i in avail_size]
    total_sizes.append(size_text)
        
        
    
    # extracting Review======================================================================================
    try:
        review_ratings =driver.find_element(By.XPATH,'//div[@class="pr-snippet-stars-reco-inline pr-snippet-standard"]/div[1]')
        
        #Extracting Reviews........................
        try:
            rev=review_ratings.find_element(By.XPATH,'.//div[@class="pr-snippet-read-and-write"]/span')
            reviews.append(rev.text.split(' ')[0])
        except NoSuchElementException:
            reviews.append('NA')
            
    except NoSuchElementException:
        reviews.append('NA')
        
    #Extracting Ratings
    try:
        rat= driver.find_element(By.XPATH,'//div[@class="pr-snippet-stars pr-snippet-stars-png "][1]/div[2][1]')
        ratings.append(rat.text)
    except NoSuchElementException:
        ratings.append('NA')
    
    #Extracting features==============================================
    
    try:
        featr = driver.find_element(By.XPATH,'//div[@id="collapsible-details-1"]/div[2]/ul')
        features.append(featr.text)
    except NoSuchElementException:
        features.append('NA')
        
    # Extracting Comfort=======================================================
    try:
        comf = driver.find_element(By.XPATH,'//div[@class="col-sm-12 col-md-8 col-lg-9 value content pl-0 details-page-long-description"]/div[1]')
        comfort.append(comf.text)
    except NoSuchElementException:
        comfort.append('NA')
```
