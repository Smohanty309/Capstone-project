### Extracting first 459

```python
new =products[:459]
```

```python
name = []
category =[]
colors = []
price =[]
count_sizes = []
sizes_list = []
color_name = []
product_code =[]
features =[]
comfort =[]

for i in new:
    driver.get(i)
    
    #Extracting Category:========================================================
    if 'referrer-category=' in i:
        category_part = i.split('referrer-category=')[-1].split('&')[0]
        category.append(category_part)
    else:
        category.append('NA')
        
    #Extracting Name:===========================================================
    try:
        title= driver.find_element(By.XPATH,'//div[@class="tw-1uc1ku0 tw-1p4ksvz"]/h1')
        name.append(title.text)
    except NoSuchElementException:
        name.append('NA')
    
    
    #Extracting Price:===================================================================
    try:
        cost = driver.find_element(By.XPATH,'//div[@data-test-id="pdp-price"]')
        
        try:
            reduced = cost.find_element(By.XPATH,'.//span[@data-test-id="item-sale-price-pdp"]')
            price.append(reduced.text.split('₹')[1])
        except NoSuchElementException:
            
            try:
                original = cost.find_element(By.XPATH,'.//span[@data-test-id="item-price-pdp"]')
                price.append(original.text.split('₹')[1])
            except NoSuchElementException:
                ('NA')
                
    except NoSuchElementException:
        price.append('NA')
        
        
    # Extracting Count of colors==================================================================
    try:
        colrs =driver.find_element(By.XPATH,'//div[@id="style-picker"]')
        count = len(colrs.find_elements(By.XPATH,'//label[@data-test-id="color"]'))
        colors.append(count)
        
    except NoSuchElementException:
        colors.append('NA')
        
        
    # Extracting colors Name:==============================================================
    try:
        col = driver.find_element(By.XPATH,'//div[@class="tw-xf6gca tw-1p4ksvz"]/h4')
        color_name.append(col.text)
        
    except NoSuchElementException:
        color_name.append('NA')
        
    #Extracting Sizes and count of all the sizes=====================================================
    try:
        size_selector =driver.find_element(By.XPATH,'//div[@id="size-picker"]')

        #==========Count_sizes=====================
        try:
            count = len(size_selector.find_elements(By.XPATH,'//label[@data-disabled="false"]'))
            count_sizes.append(count)
        except NoSuchElementException:
            count_sizes.append('NA')

        #Extracting Sizes============================
        try:
            size_text=size_selector.find_elements(By.XPATH,'.//span[@data-content="size-value"]')
            sizes=[i.text.split(' ')[1] for i in size_text]
            sizes_list.append(sizes)

        except NoSuchElementException:
            sizes_list.append('NA')
            
    except NoSuchElementException:
        count_sizes.append('NA')
        sizes_list.append('NA')
    
    #Scroll=================================================
    try:
        scroll=driver.find_element(By.XPATH,'//div[@data-test-id="pdp-product-description"]')
        driver.execute_script("arguments[0].scrollIntoView(true);", scroll)
    except NoSuchElementException:
        pass
    
    # Extracting Product_code=======================================================
    try:
        code = driver.find_element(By.XPATH,'//ul[@class="tw-1h4nwdw tw-p9uz4a tw-xwzea6 list-disc list-inside"]/li[1]')
        product_code.append(code.text.split(':')[1].strip())
        
    except NoSuchElementException:
        product_code.append('NA')
        
    #Scroll till Product Description========================================================
    try:
        scroll1 = driver.find_element(By.XPATH,'//section[@id="product-story"]')
        driver.execute_script("arguments[0].scrollIntoView(true);", scroll1)
    except NoSuchElementException:
        pass
    
    #Extracting Comfort===============================================================
    try:
        comfi = driver.find_element(By.XPATH,'//section[@class="grid gap-8 lg:gap-10 grid-cols-1 md:grid-cols-2"]/div[2]/ul')
        comfort.append(comfi.text)
    except NoSuchElementException:
        comfort.append('NA')
        
    # Extracting features======================================================================
    try:
        des = driver.find_element(By.XPATH,'//section[@class="grid gap-8 lg:gap-10 grid-cols-1 md:grid-cols-2"]/div[3]/ul')
        features.append(des.text)
    except NoSuchElementException:
        features.append('NA')
```

```python
len(name), len(category), len(price), len(colors), len(color_name), len(count_sizes), len(sizes_list), len(product_code), len(comfort), len(features)
```

```python
#Converting in dataframe

puma_1sturls=pd.DataFrame({
    'Name': name,
    'Category': category,
    'Colors': colors,
    'Price': price,
    'Sizes': sizes_list,
    'Count_Sizes': count_sizes,
    'Color_Name': color_name,
    'product_Code': product_code,
    'Features' : features,
    'Comfort' : comfort,
    
})

puma_1sturls
```

```python
puma_1sturls.to_csv('puma_1sturls.csv', index=True)
```
